```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import color
from skimage import io

imagen_original = plt.imread("prueba22.jpg")
plt.imshow(imagen_original)
plt.show()

imagen = color.rgb2gray(imagen_original)
plt.imshow(imagen, cmap = 'gray')
plt.show()

print(imagen.shape)

ancho = imagen.shape[0]
largo = imagen.shape[1]

print(ancho)
print(largo)

print(imagen[100][200])
print(imagen[200][300])
print(imagen[250][600])

dx=np.zeros(shape=(ancho, largo))

for i in range (0, ancho-1):
    for j in range(0, largo-1):
        if 1-abs(imagen[i+1][j] - imagen[i][j]) < 0.95:
            dx[i][j]= 1
        else:
            dx[i][j]=0
                
plt.imshow(dx, cmap = 'gray')
plt.show()

dy=np.zeros(shape=(ancho, largo))

for i in range(0, ancho-1):
    for j in range (0, largo-1):
        if 1-abs(imagen[i][j+1] - imagen[i][j]) < 0.95:
            dy[i][j]= 1
        else:
            dy[i][j]=0
                
plt.imshow(dy,cmap = 'gray')
plt.show()

gradiente = np.zeros(shape=(ancho, largo))
for i in range(0, ancho):
    for j in range(0, largo):
        gradiente[i][j] = dx[i][j] + dy[i][j]
        
plt.imshow(gradiente , cmap = 'gray')
plt.show()

for i in range(0, ancho):
    for j in range(0, largo):
        gradiente[i][j] = 1 - gradiente[i][j]
        
plt.imshow(gradiente, cmap = 'gray')
plt.show()

diagonalD=np.zeros(shape=(ancho, largo))

for i in range(0, ancho-1):
    for j in range (0, largo-1):
        if 1-abs(imagen[i+1][j+1] - imagen[i][j]) < 0.95:
            diagonalD[i][j]= 0
        else:
            diagonalD[i][j]=1
           
plt.imshow(diagonalD,cmap = 'gray')
plt.show()

diagonalI=np.zeros(shape=(ancho, largo))

for i in range(0, ancho-1):
    for j in range (0, largo-1):
        if 1-abs(imagen[i][j] - imagen[i-1][j-1]) < 0.95:
            diagonalI[i][j]= 0
        else:
            diagonalI[i][j]=1
            
plt.imshow(diagonalI,cmap = 'gray')
plt.show()

sumaDiagonales = np.zeros(shape=(ancho, largo))
for i in range(0, ancho):
    for j in range(0, largo):
        sumaDiagonales[i][j] = diagonalD[i][j] + diagonalI[i][j]
        
plt.imshow(sumaDiagonales , cmap = 'gray')
plt.show()
```
